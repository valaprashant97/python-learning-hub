### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a Leap Year")
else:
    print(f"{year} is not a Leap Year")
```

### 07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

choice = input("Enter +,-,*,\ :")

match(choice):
    case '+':
        print(a+b)
    case '-':
        print(a-b)
    case '*':
        print(a*b)
    case '/':
        print(a/b)
    case _:
        print("invalide input")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
s1 = int(input("Enter a marks subject 1 :"))
s2 = int(input("Enter a marks subject 2 :")) 
s3 = int(input("Enter a marks subject 3 :")) 
s4 = int(input("Enter a marks subject 4 :")) 
s5 = int(input("Enter a marks subject 5 :")) 

avg_marks = (s1+s2+s3+s4+s5)/5
if avg_marks < 35:
    print("Fail")
elif 35 <= avg_marks < 45:
    print("Pass Class")
elif 45 <= avg_marks < 60:
    print("Second Class")
elif 60 <= avg_marks < 70:
    print("First Class")
else:
    print("Distiction")
```

### 10) WAP to calculate electricity bill based on following criteria. Which takes the unit from the user.
#### a. First 1 to 50 units – Rs. 2.60/unit
#### b. Next 50 to 100 units – Rs. 3.25/unit
#### c. Next 100 to 200 units – Rs. 5.26/unit
#### d. above 200 units – Rs. 8.45/unit

```python
units = float(input("Enter total units consumed: "))

if units <= 50:
    bill = units * 2.60
elif units <= 100:
    bill = (50 * 2.60) + ((units - 50) * 3.25)
elif units <= 200:
    bill = (50 * 2.60) + (50 * 3.25) + ((units - 100) * 5.26)
else:
    bill = (50 * 2.60) + (50 * 3.25) + (100 * 5.26) + ((units - 200) * 8.45)
print(f"Total Units: {units}")
print(f"Total Bill: Rs. {bill:.2f}")
```
