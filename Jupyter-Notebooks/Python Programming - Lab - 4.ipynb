### 10) WAP to print GCD of given two numbers.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

gcd = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i

print("gcd is:", gcd)
```

### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input("Enter a number: "))

count = 0
for i in range(1, n + 1):
    if n % i == 0:
        count += 1

if count == 2:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

### 13) WAP to print sum of digits of given number.

```python
n = int(input("Enter a number: "))

sum = 0
for i in range(len(str(n))):
    digit = n % 10
    sum = sum + digit
    n = n // 10

print("Sum of digits:", sum)
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter a number: "))

num = n
rev = 0

for i in range(len(str(n))):
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

if num == rev:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n = int(input("Enter a number: "))

num = n
digits = len(str(n))
sum = 0

for i in range(digits):
    digit = n % 10
    sum = sum + digit ** digits
    n = n // 10

if sum == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n = int(input("Enter a number: "))

for i in range(1, n + 1):
    sum = 0
    for j in range(1, i):
        if i % j == 0:
            sum = sum + j
    if sum == i:
        print(i)
```
