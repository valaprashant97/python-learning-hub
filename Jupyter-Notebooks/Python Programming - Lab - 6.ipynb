### 04) WAP to interchange first and last elements in a list.

```python
mylist = [1, 2, 3, 4, 5, 6]
length = 0
for i in mylist:
    length = length + 1
    
tamp = mylist[0]
mylist[0] = mylist[length-1]
mylist[length-1] = tamp

print(mylist)
```

### 05) WAP to split the List into two parts and append the first part to the end.

```python
mylist = [1, 2, 3, 4, 5, 6]
n = 3
result = []

for i in range(n, len(mylist)):
    result.append(mylist[i])

for i in range(n):
    result.append(mylist[i])

print(result)
```

### 09) WAP to count unique items in a list.

```python
n = int(input("Enter number of elements: "))
lst = []

for i in range(n):
    lst.append(int(input("Enter element: ")))

unique_list = []
for i in range(n):
    if lst[i] not in unique_list:
        unique_list.append(lst[i])

count = len(unique_list)
print(count)
```

### 13) WAP to find second largest number in a list.

### 14) WAP to extract elements with frequency greater than K.

```python
n = int(input("Enter number of elements: "))
lst = []

for i in range(n):
    lst.append(int(input("Enter element: ")))
k = int(input("Enter a k :"))
res = []
for i in lst:
    if lst.count(i) > k and i not in res:
        res.append(i)
print(res)
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
n = int(input("Enter number of elements: "))
lst = []

for i in range(n):
    lst.append(input("Enter element: "))
ans = []
for i in lst:
    if i[0] == 'b':
        ans.append(i)
print(ans)
```

### 17) WAP to create a list of common elements from given two lists.

```python
n1 = int(input("Enter number of elements in list 1: "))
lst1 = []

for i in range(n1):
    lst1.append(int(input("Enter element: ")))

n2 = int(input("Enter number of elements in list 2: "))
lst2 = []

for j in range(n2):
    lst2.append(int(input("Enter element: ")))

ans = []

for i in lst1:
    for j in lst2:
        if i == j:
            ans.append(i)

print(ans)
```
