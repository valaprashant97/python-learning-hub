### 01) WAP to find their union, intersection, and difference of the given two sets.
**Sample Input:**
```
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

```
**Sample Output:**
```
Union: {1, 2, 3, 4, 5, 6}
Intersection: {3, 4}
Difference (A-B): {1, 2}

```

````python
s1 = set(map(int, input("Enter elements of set1: ").split(",")))
s2 = set(map(int, input("Enter elements of set2: ").split(",")))

union= s1| s2
intersection= s1 & s2
differance= s1-s2

print("Union:",union)
print("Intersection:",intersection)
print("Difference (s1-s2):",differance)
````

### 03) WAP to remove duplicate elements from a list while preserving unique values.
**Sample Input:**
```
[10, 20, 10, 30, 20, 40]

```
**Sample Output:**
```
{10, 20, 30, 40}
```

````python
l1 = list(map(int, input("Enter elements of list: ").split(",")))
l2=[]
for i in l1:
    if i not in l2:
        l2.append(i)
l2 = set(l2)
print("After Remove Duplicate Elements:",l2)
````

### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
l1 = list(map(int, input("Enter elements of set1: ").split(",")))
l2 = list(map(int, input("Enter elements of set1: ").split(",")))
l3 = list(map(int, input("Enter elements of set2: ").split(",")))
l4 = list(map(int, input("Enter elements of set2: ").split(",")))
Dict1 = dict(zip(l1,l2))
Dict2 = dict(zip(l3,l4))
merged = {}

for key in Dict1:
    merged[key] = Dict1[key]

for key in Dict2:
    if key in merged:
        merged[key] += Dict2[key]
    else:
        merged[key] = Dict2[key]

print("Merged Dictionary:", merged)
````

### 06) WAP to Sort a dictionary in ascending order of its values.
Hint: in sorted(), use udf for key parameter


**Sample Input:**
```
{'Math': 70, 'Physics': 85, 'Chemistry': 60}

```
**Sample Output:**
```
{'Chemistry': 60, 'Math': 70, 'Physics': 85}
```

````python
def sort_by_value(item):
    return item[1]

d = {'Math': 70, 'Physics': 85, 'Chemistry': 60}

sorted_dict = dict(sorted(d.items(), key=sort_by_value))

print(sorted_dict)
````

### 07) WAP to find the keys having maximum and minimum values in a dictionary.

**Sample Input:**
```
{'A': 50, 'B': 90, 'C': 30}
```
**Sample Output:**
```
Max Value Key: B
Min Value Key: C

```

````python
d = {'A': 50, 'B': 90, 'C': 30}

max_key = max(d, key=d.get)

min_key = min(d, key=d.get)

print("Max Value Key:", max_key)
print("Min Value Key:", min_key)
````

### 09) WAP to count the frequency of each character in a string using a dictionary.
**Sample Input:**
```
programming
```
**Sample Output:**
```
{'p': 1, 'r': 2, 'o': 1, 'g': 2, 'a': 1, 'm': 2, 'i': 1, 'n': 1}
```

````python
s = input("Enter a string: ")

freq = {}

for ch in s:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)      
````

### 11) WAP to find the student with the highest total marks, given student names and their marks.

**Sample Input:**
```
{'Amit': [70, 80, 90], 'Riya': [85, 75, 80], 'Neha': [60, 70, 65]}
```
**Sample Output:**
```
Topper: Amit
```

````python
students = {
    'Amit': [70, 80, 90],
    'Riya': [85, 75, 80],
    'Neha': [60, 70, 65]
}

max_total = 0
topper = ""

for name, marks in students.items():
    total = sum(marks)
    if total > max_total:
        max_total = total
        topper = name

print("Topper:", topper)
````

### 12) WAP to extract keys whose values are even numbers.

**Sample Input:**
```
{1: 11, 2: 20, 3: 15, 4: 40}
```
**Sample Output:**
```
[2, 4]
```

````python
d = {1: 11, 2: 20, 3: 15, 4: 40}

even_keys = []

for key, value in d.items():
    if value % 2 == 0:
        even_keys.append(key)

print(even_keys)
````

### 15) WAP to create a dictionary using dictionary comprehension from two lists containing student names and marks. Include only those students in the dictionary whose marks are greater than or equal to 75.

**Sample Input:**
```
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]
```
**Sample Output:**
```
{'Amit': 78, 'Neha': 85}
```

````python
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]

result = {students[i]: marks[i] for i in range(len(students)) if marks[i] >= 75}

print(result)
````
