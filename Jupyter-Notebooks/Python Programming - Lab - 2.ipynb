### 13) WAP to calculate area and perimeter of a circle.

```python
import math

r = float(input("Enter radius of the circle: "))

area = math.pi * r * r
perimeter = 2 * math.pi * r

print(f"Area of circle: {area:.4f}")
print(f"Perimeter of circle: {perimeter:.4f}")
```

### 14) WAP to convert degree into Fahrenheit and vice versa.

```python
temp = float(input("Enter temperature: "))
choice = input("Enter C to convert to Fahrenheit or F to convert to Celsius: ")

if choice == 'C' or choice == 'c':
    fahrenheit = (temp * 9/5) + 32
    print(fahrenheit)
elif choice == 'F' or choice == 'f':
    celsius = (temp - 32) * 5/9
    print(celsius)
else:
    print("Invalid choice")
```

### 15) WAP to get the distance from user into kilometer, and convert it into meter, feet, inches and centimeter.
