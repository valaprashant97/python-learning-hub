### 03) WAP to find tuples which have all elements divisible by K from a list of tuples.
**Sample Input:**
```
tuple_list = [(2, 4, 6), (3, 6, 9), (4, 8, 12), (5, 10, 15)]
K = 2

```
**Sample Output:**
```
Tuples divisible by 2 : [(2, 4, 6), (4, 8, 12)]
```

````python
tuple_list = [(2,4,6), (3,6,9), (4,8,12), (5,10,15)]
K = 2
res = [t for t in tuple_list if all(x % K == 0 for x in t)]
print("Tuples divisible by", K, ":", res)
````

### 14) WAP to create (index, value) pairs for prime numbers.

**Sample Input:**
```
T = (4, 7, 9, 11, 15)
```
**Sample Output:**
```
((1, 7), (3, 11))
```

````python
T = tuple(map(int, input().split()))

result = []

for i in range(len(T)):
    n = T[i]
    if n > 1:
        prime = True
        for j in range(2, n):
            if n % j == 0:
                prime = False
                break
        if prime:
            result.append((i, n))

print(tuple(result))
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
T = eval(input())

result = [sum(col) for col in zip(*T)]

print(*result)
````
